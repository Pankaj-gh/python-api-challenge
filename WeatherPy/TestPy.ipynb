```python
#Finding random 503 integers
randoms=np.random.randint(0,total,503)
```

```python
#Creating a new DataFrame to add random 500 rows from the original DataFrame
new_city= pd.DataFrame({"Country":"","City":"","Latitude":0, "Longitude":0},index=[0])
```

```python
new_city
```

```python
#Looping over those 503 random integrs to get an completely random DataFrame 
for index,rows in cities.iterrows():
    for nos in randoms:
        new_city.loc[nos,"Country"]=cities.loc[nos,"Country"]
        new_city.loc[nos,"City"]=cities.loc[nos,"City"]
        new_city.loc[nos,"Latitude"]=cities.loc[nos,"Latitude"]
        new_city.loc[nos,"Longitude"]=cities.loc[nos,"Longitude"]
    break
    
```
